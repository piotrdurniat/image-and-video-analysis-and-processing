Przykłady detekcji krawędzi (Sobel, Canny) w OpenCV

```python
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img = cv2.imread("Lena.jpg", cv2.IMREAD_GRAYSCALE)
Image.fromarray(img)
```

Detekcja krawędzi metodą Sobela

```python
# w kierunku X:
sobelx = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=3)

# w kierunku Y:
sobely = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=3)

# w kierunkach X i Y jednocześnie - doczytaj w dokumentacji, co dokładnie jest wynikiem tej operacji:
sobelxy = cv2.Sobel(src=img, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=3)

f, (a1, a2, a3) = plt.subplots(1, 3)
f.set_figwidth(15)
f.set_figheight(5)
a1.imshow(sobelx, cmap="gray")
a1.set_title("krawędzie o orientacji pionowej")
a2.imshow(sobely, cmap="gray")
a2.set_title("krawędzie o orientacji poziomej")
a3.imshow(sobelxy, cmap="gray")
a3.set_title("krawędzie o orientacjach pionowej i pionowej")
```

Detekcja krawędzi metodą Canny'ego

```python
# wynik zależy od doboru wartości progów
f, ((a1, a2, a3), (a4, a5, a6)) = plt.subplots(2, 3)
f.set_figwidth(15)
f.set_figheight(10)

a1.imshow(img, cmap="gray")
a1.set_title("obraz oryginalny")

for ax, t1, t2 in zip(
    [a2, a3, a4, a5, a6], [50, 50, 250, 20, 20], [200, 300, 300, 100, 300]
):
    ax.imshow(cv2.Canny(image=img, threshold1=t1, threshold2=t2), cmap="gray")
    ax.set_title("Canny, próg dolny = {}, próg górny = {}".format(t1, t2))
```
